## 3. Sử dụng LLM với LangChain

Đoạn mã dưới đây cho phép chúng ta tương tác với mô hình tinyllama được lưu trữ cục bộ trên máy tính của chúng ta. Chúng ta chỉ đơn giản gửi thông điệp sau đến mô hình: "I love programming".

Vui lòng đọc các bình luận trong mã nếu bạn muốn hiểu chính xác điều gì đang diễn ra.

```python
# Import the libraries
from langchain_ollama import ChatOllama
from langchain_core.messages import AIMessage

# Initialize the LLM. This is when we pick the model and the temperature (which controls the randomness of the output). llm is what we will use to call the model.
llm = ChatOllama(
    model="tinyllama:1.1b",
    temperature=0,  # 0 for more deterministic outputs
)

# We need to provide an array of messages to the model. The first message is always the system message, which tells the model what it is. The second message is the human message, which is what we want to ask the model.
messages = [
    (
        "system",
        "You are a helpful assistant",
    ),
    ("human", "I love programming."),
]

# Send the array of messages to the model and get the response. The response is an AIMessage object, which contains the content of the message. Which we will print below.
ai_msg = llm.invoke(messages)

# Display the response
print(ai_msg.content)
```

## 5. Tạo một Chatbot Cơ bản

Hãy tạo một giao diện trò chuyện đơn giản để chúng ta có thể tương tác với mô hình. Mỗi ô sẽ đại diện cho một tương tác. Theo dõi các bước dưới đây, chúng sẽ có ý nghĩa hơn khi bạn thực hiện chúng.

```python
# Setup: Run this cell first
from langchain_ollama import ChatOllama

# Initialize the LLM
llm = ChatOllama(
    model="tinyllama:1.1b",
    temperature=0
)

# System prompt
system_prompt = "You are a helpful assistant"

print("-" * 30)
```

Bây giờ, hãy tạo một ô cho tương tác đầu tiên của chúng ta bằng cách nói với mô hình màu sắc yêu thích của chúng ta là gì. Lưu ý rằng phản hồi có thể kỳ lạ, một lần nữa chúng ta không sử dụng mô hình tốt nhất hiện có. Nhưng nó sẽ cho thấy rằng nó đã hiểu rằng màu sắc yêu thích của bạn là màu xanh lam.

```python
# First interaction 
user_message = "My favorite color is blue."

# Create fresh messages for this interaction only
messages = [
    ("system", system_prompt),
    ("human", user_message)
]

# Get response
ai_msg = llm.invoke(messages)

print(f"You: {user_message}")
print(f"Assistant: {ai_msg.content}")
```

Bây giờ hãy tiếp tục với một câu hỏi:

## 6. Hiểu cách hoạt động của bộ nhớ hội thoại

Cách dễ nhất để quản lý điều đó là chỉ cần truyền toàn bộ lịch sử cuộc trò chuyện cho mô hình, thay vì chỉ từng tin nhắn riêng lẻ:

```python
from langchain_ollama import ChatOllama
from langchain.schema import SystemMessage, HumanMessage

# Initialize the LLM like before
llm = ChatOllama(
    model="tinyllama:1.1b",
    temperature=0
)

# Define the conversation history (always starts with a system prompt)
conversation = [SystemMessage(content="You are a helpful assistant")] #conversation is an array (list) of messages. SystemMessage is a class that represents a system message and accepts content as a parameter.

#Define a function called chat which takes user input as a parameter. This function will be used to interact with the model.
def chat(user_input: str):
    # 1) Append the user's message
    conversation.append(HumanMessage(content=user_input)) #We add the user message to the conversation list. Now we have the system message and the user message in the conversation list.
    
    # 2) Call the model with the full conversation list (user message + system message)
    ai_msg = llm(conversation)
    
    # 3) Display the exchange
    print(f"You: {user_input}")
    print(f"Assistant: {ai_msg.content}")
    print(f"[Conversation length: {len(conversation) + 1} messages]")  # +1 for the AI response
    
    # 4) Add the AI's response to history. So now we have the system message, user message, and AI response in the conversation list which will be used for the next interaction.
    conversation.append(ai_msg)
    
    return ai_msg.content

print("Chat With Memory")
print("-" * 30)
```

Hãy gọi lại màu yêu thích của chúng ta:

```python
# First interaction - share favorite color
chat("My favorite color is blue")
```

Bây giờ hãy xem liệu nó có nhớ không:

```python
# Second interaction - test memory
chat("What's my favorite color?")
```

Giờ đây mô hình sẽ ghi nhớ tên của bạn một cách chính xác! Điều này là do chúng ta đang theo dõi lịch sử cuộc trò chuyện.

## 7. Thách thức về Cửa sổ ngữ cảnh (Context Window)

Mặc dù giải pháp bộ nhớ đơn giản của chúng ta hoạt động ban đầu, bởi vì chúng ta chỉ truyền toàn bộ lịch sử tin nhắn làm đầu vào, nhưng nó có một hạn chế nghiêm trọng: **cửa sổ ngữ cảnh (context window)**.

### Hiểu về Cửa sổ ngữ cảnh (Context Window)

Mỗi mô hình ngôn ngữ có một "cửa sổ ngữ cảnh" cố định - số lượng token tối đa (gần đúng là từ hoặc các phần của từ) mà nó có thể xử lý cùng một lúc:

- **TinyLlama**: ~2.048 token (khoảng 1.500 từ)
- **GPT-3.5**: ~4.096 token
- **GPT-4**: ~8.192 đến 128.000 token tùy thuộc vào phiên bản
- **Claude 3**: Lên đến 200.000 token

Khi cuộc trò chuyện phát triển, cuối cùng chúng ta sẽ đạt đến giới hạn này. Khi điều đó xảy ra:

1. Mô hình không thể nhìn thấy các tin nhắn ngoài kích thước cửa sổ
2. Mô hình thực sự "quên" những phần đầu tiên của cuộc trò chuyện
3. Quá trình xử lý trở nên chậm hơn với các ngữ cảnh lớn hơn
4. Bạn có thể nhận được lỗi nếu bạn vượt quá cửa sổ ngữ cảnh

## 8. Các giải pháp cho vấn đề cửa sổ ngữ cảnh (Context Window)

Trong các hệ thống sản xuất như ChatGPT, một số kỹ thuật giải quyết hạn chế về cửa sổ ngữ cảnh:

### 1. Cửa sổ ngữ cảnh trượt (Sliding Context Window)
Cách tiếp cận này chỉ giữ lại N tin nhắn gần đây nhất, cộng với lời nhắc hệ thống.

**Ví dụ:** Nếu bạn giới hạn ở 10 tin nhắn và cuộc trò chuyện của bạn đạt 15 tin nhắn, bạn sẽ bỏ 5 tin nhắn cũ nhất (ngoại trừ lời nhắc hệ thống).

Điều này giống như việc tham gia một cuộc trò chuyện mà bạn chỉ nhớ những gì đã được nói trong vài phút gần đây. Nó đơn giản nhưng làm mất tất cả thông tin cũ hơn.

### 2. Tóm tắt (Summarization)
Kỹ thuật này cô đọng các phần cũ hơn của cuộc trò chuyện thành các bản tóm tắt để tiết kiệm không gian token.

**Ví dụ:** Sau 10 lần trao đổi qua lại, hệ thống có thể thay thế 20 tin nhắn đó bằng một bản tóm tắt duy nhất: "Người dùng giới thiệu bản thân là Alex và hỏi về mạng nơ-ron. Trợ lý giải thích về mạng nơ-ron và cung cấp một ví dụ mã Python để phát hiện chuỗi đối xứng."

Điều này giữ lại các điểm chính trong khi giảm đáng kể việc sử dụng token.

### 3. Quản lý phiên được hỗ trợ bởi cơ sở dữ liệu (Database-Backed Session Management)

Cách tiếp cận này sử dụng cơ sở dữ liệu để lưu trữ lịch sử cuộc trò chuyện hoàn chỉnh được liên kết với các phiên người dùng duy nhất. Đây là cách hoạt động:

1. **Tạo phiên**: Khi người dùng bắt đầu trò chuyện, hệ thống tạo một ID phiên duy nhất (ví dụ: "session_abc123")

2. **Lưu trữ tin nhắn**: Mọi tin nhắn từ người dùng và AI đều được lưu trữ trong bảng cơ sở dữ liệu được liên kết với ID phiên đó

3. **Quản lý cửa sổ ngữ cảnh**: Khi chuẩn bị lời nhắc cho LLM, hệ thống:
   - Truy xuất tất cả tin nhắn cho phiên
   - Áp dụng các chiến lược để phù hợp với cửa sổ ngữ cảnh (như cửa sổ trượt hoặc tóm tắt)
   - Gửi cuộc trò chuyện đã tối ưu hóa đến mô hình
   - Lưu trữ phản hồi mới trở lại cơ sở dữ liệu

**Ví dụ trong thực tế:**
- Một người dùng đã trò chuyện hàng giờ với ChatGPT
- ID phiên của họ "user_789" hiện có 200 tin nhắn trong cơ sở dữ liệu
- Khi họ gửi tin nhắn #201, hệ thống:
  - Truy xuất tất cả 200 tin nhắn trước đó từ cơ sở dữ liệu
  - Chọn những tin nhắn quan trọng nhất để phù hợp với cửa sổ ngữ cảnh
  - Nhận phản hồi từ mô hình
  - Thêm tin nhắn #201 và phản hồi vào cơ sở dữ liệu
  - Ngay cả khi mô hình chỉ "thấy" phần gần đây, toàn bộ lịch sử vẫn còn trong cơ sở dữ liệu

Đây là cách các dịch vụ như ChatGPT có thể duy trì "bộ nhớ" trong các cuộc trò chuyện rất dài và ngay cả khi bạn đóng trình duyệt và quay lại sau.

## Những gì ChatGPT thường sử dụng

ChatGPT sử dụng một cách tiếp cận kết hợp nhiều kỹ thuật này:

1. **Kỹ thuật chính:** Nó sử dụng một cửa sổ ngữ cảnh rất lớn (lên đến 128K token trong GPT-4o) và quản lý phiên được hỗ trợ bởi cơ sở dữ liệu, cho phép nó "nhớ" các cuộc trò chuyện dài đáng kể.

2. **Đối với các cuộc trò chuyện dài:** Khi các cuộc trò chuyện vượt quá cả những giới hạn hào phóng này, nó sử dụng các kỹ thuật cửa sổ trượt ưu tiên:
   - Lời nhắc/hướng dẫn hệ thống
   - Các tin nhắn gần đây nhất
   - Các tin nhắn có mật độ thông tin cao
   - Các tin nhắn được người dùng tham chiếu rõ ràng

3. **Nén động:** Trong một số phiên bản, ChatGPT cũng sử dụng các thuật toán nén động chọn lọc tóm tắt hoặc loại bỏ các phần của cuộc trò chuyện có vẻ ít liên quan đến cuộc trao đổi hiện tại.

Mặc dù các giải pháp này giúp ích, nhưng vẫn có một giới hạn tuyệt đối về những gì bất kỳ mô hình nào có thể "nhớ" trong một cuộc trò chuyện duy nhất. Đây là lý do tại sao ngay cả ChatGPT đôi khi "quên" những điều được đề cập sớm hơn nhiều trong các cuộc trò chuyện rất dài.
